```python
lst=[23,56,17,9,56,42,79,26,18]
print('Original list:\n',lst)
ans='Y'
while ans=='y' or ans=='Y':
  print('LIST OPERATIONS')
  print('1. Add an item at end of list')
  print('2. Add an item at specified index position of list')
  print('3. Remove a specified element')
  print('4. Remove an element from a specified position')
  print('5. Reverse the elements of list')
  print('6. Display')
  ch= int(input('Enter your choice'))
  if ch==1:
    item=int(input('Enter item to be added'))
    lst.append(item)
    print(item,'append')
  elif ch==2:
    item=int(input('Enter item to be added at index'))
    pos=int(input('Enter index position'))
    lst.insert(pos,item)
    print(item,'added at index position',pos)
  elif ch==3:
    item=int(input('Enter item to be removed'))
    lst.remove(item)
    print(item,'removed')
  elif ch==4:
    pos=int(input('Enter index position of item to be removed'))
    ele=lst.pop(pos)
    print(ele,'removed from index position',pos)
  elif ch==5:
    lst.reverse()
    print('List reversed')
  elif ch==6:
    print('The list is \n',lst)
  else:
    print('Invalid choice')
  ans=input('Do you wish to continue(Y/N)')
```
