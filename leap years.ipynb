```python
start = int(2020)
list1=[]
end = int(input("Enter end year:"))
while start < end:
    if start % 4 == 0:
       if start % 100 == 0:
          if start % 400 == 0:
              list1.append(start)
       else:
           list1.append(start)
    start += 1
print("List of leap years in the range is :",list1)
```
