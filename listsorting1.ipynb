```python
lst=eval(input('Enter a list of integers'))
oddlst=[]
evenlst=[]
oddcount=0
evencount=0
for a in lst:
   if a%2==0:
     evenlst.append(a)
     evencount += 1
   else:
     oddlst.append(a)
     oddcount += 1
evenlst.sort()
oddlst.sort()
print('Original list',lst)
print('Even list in ascending order',evenlst)
print('Odd list in ascending order',oddlst)
if evencount>oddcount:
  print('There are more even elements')
elif oddcount>evencount:
  print('There are more odd elements')
else:
  print('There are equal no: of even and odd elements')
```
